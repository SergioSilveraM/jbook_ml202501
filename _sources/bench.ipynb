# ***Bench***

```python
from reporte_metricas import ReporteMetricas
reporte = ReporteMetricas()
df = reporte.load()
```

## ***Validation***

### ***macro avg***

```python
dfval_macro = df.query("Type == 'val' and Class == 'macro avg'").drop(columns=["accuracy", "log_loss", "auc"]).reset_index(drop=True)
dfval_macro.sort_values(by=["f1-score"], ascending=False, inplace=True)
dfval_macro
```

### ***weighted avg***

```python
dfval_weig = df.query("Type == 'val' and Class == 'weighted avg'").drop(columns=["accuracy", "log_loss", "auc"]).reset_index(drop=True)
dfval_weig.sort_values(by=["f1-score"], ascending=False, inplace=True)
dfval_weig
```

### ***Global***

Ojo que los AUC y logloss para SVC por el algoritmo utilizado no fueron calculados

```python
dfval_global = df.query("Type == 'val' and Class == 'global'").drop(columns=["precision", "recall", "f1-score", "support"]).reset_index(drop=True)
dfval_global.sort_values(by=["auc"], ascending=False, inplace=True)
dfval_global
```
